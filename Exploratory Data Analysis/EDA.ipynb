```python
# ─── 1) Imports & Data Loading ────────────────────────────────────────────────
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# fetch 3 years of daily data
ticker = "7010.SR"
df = yf.download(ticker, period="5y", progress=False)
```

```python
## 2) Close‐Price Trend
```

```python

plt.figure(figsize=(12, 5))

# extract as plain numpy arrays
x = df.index.to_numpy()             # array of Timestamps
y = df["Close"].to_numpy().flatten()  # 1-D float array

plt.plot(x, y, label="Close Price", linewidth=1.5)
plt.title("STC (7010.SR) Close Price Over 5 Years")
plt.xlabel("Date")
plt.ylabel("Close (SAR)")
plt.legend()
plt.tight_layout()
plt.savefig("close_trend.png", dpi=150)
plt.show()
```
